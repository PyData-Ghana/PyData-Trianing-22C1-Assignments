```python

from selenium import webdriver
#import webdriver from selenium
```

```python
url = 'https://www.jumia.com.gh/laptops/'
#the url to scrape the info
driver = webdriver.Chrome()
#open chrome using webdriver
```

```python
driver.get(url)
#open url using chrome webdriver
```

```python
laptop_names = [names.text for names in driver.find_elements_by_xpath('//div[@class="info"]//child::h3[@class="name"]')]
laptop_names
#using xpath to get all laptop names. from h3 tag 
```

```python
laptop_prices = [prices.text for prices in driver.find_elements_by_xpath('//div[@class="info"]//child::div[@class="prc"]')]
laptop_prices
```

```python
laptop_links = [links.get_attribute("href") for links in driver.find_elements_by_xpath('//a[@class="core"]')]
laptop_links

# get href using 'a' tag with class name 'core'
```

```python
import pandas as pd
#import pandas library
Laptops = pd.DataFrame({
     "Laptop Name":laptop_names,
    "Price": laptop_prices,
    "Links" : laptop_links[47]
})
Laptops
#put results into dataframe
```

```python
Laptops.to_csv("Jumia_laptops_using_selenium.csv")
```
