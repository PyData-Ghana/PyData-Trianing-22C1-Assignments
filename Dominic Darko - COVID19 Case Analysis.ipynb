```python
df = pd.read_csv('covid_data.csv')
df
```

```python
case_vaccin_df = df.loc[:, ['actuals.cases', 'actuals.vaccinationsCompleted']]
graph_covid = case_vaccin_df.fillna(0)
graph_covid.tail(50)
```

```python
# plt.plot(x = 'actuals.cases', y = 'actuals.vaccinationsCompleted', data = graph_covid )
x = graph_covid['actuals.cases']
y1 = graph_covid['actuals.vaccinationsCompleted']
plt.title('COVID Cases & Completed Vaccin Relationship')
plt.plot(x,  label = 'COVID Cases', color = 'red')
plt.plot(y1, label = 'Vaccins Completed', color = 'green')
plt.legend()
plt.xlabel('COVID Cases')
```

### From the graph it can be seen that covid cases and vacination completion were equal to a certain point.
### However, both vaccine completion and COVID cases increased but vaccine completion continued to increase while COVID cases rose marginally and remained the same.
