```python
from selenium import webdriver
```

```python
url = 'https://www.jumia.com.gh/laptops/'
driver = webdriver.Chrome()
```

```python
driver.get(url)
```

```python
laptop_names= [names.text for names in driver.find_elements_by_xpath('//h3')]
laptop_names
```

```python
laptop_prices= [prices.text for prices in driver.find_elements_by_xpath('//div[@class= "info"]//child::div[@class="prc"]')]
laptop_prices
```

```python
laptop_links= [links.get_attribute("href") for links in driver.find_elements_by_xpath('//article[contains(@class,"prd _fb")]//child::a')]
laptop_links
```

```python
import pandas as pd
# import pandas library
Laptops = pd.DataFrame({
    "Laptop Name": laptop_names,
    "Price": laptop_prices,
    "Link":laptop_links
    
})
Laptops
```

```python
Laptops.to_csv("jumia_laptops_using_selenium_xpath.csv")
```
